### Importing important Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
```

Importing Data set and making a copy of it if requires in future

here `Loan Id` , `ApplicantIncome` , `CoapplicantIncome` , `LoanAmount` , `Loan_Amount_Term` doesn't affect the prediction. Lets drop those column

```python
train_new = train.drop(['Loan_ID'],axis=1)
train_new.head()
```

```python
lb_make = LabelEncoder()
train_new['Loan_Status'] = lb_make.fit_transform(train_new['Loan_Status'])
train_new.head()
```

Gender = 2.117, Married = 0.488, Dependents = 2.44, Self_Employed =  5.21, LoanAmount = 3.583, LoanAmountTerm = 2.28, CreditHistory = 8.14

```python
train_new['Loan_Amount_Term'].value_counts()
```

```python
train_new['Gender'].fillna(train_new['Gender'].mode()[0],inplace=True)
train_new['Married'].fillna(train['Married'].mode()[0],inplace=True)
train_new['Dependents'].fillna(train_new['Dependents'].mode()[0],inplace=True)
train_new['Self_Employed'].fillna(train_new['Self_Employed'].mode()[0],inplace=True)
train_new['Credit_History'].fillna(train_new['Credit_History'].mode()[0],inplace=True)
train_new['LoanAmount'].fillna(train_new['LoanAmount'].mean(),inplace=True)
train_new['Loan_Amount_Term'].fillna(train_new['Loan_Amount_Term'].mean(),inplace=True)
```

```python
train_new.isnull().any()
```

```python
train_new.head()
```

```python
lr = LogisticRegression()
```

```python
X_train = train_new.iloc[:,:-1]
X_train.head()
```

```python
X_train.dtypes
```

```python
def trans(df,col):
    try:
        df[col] = lb_make.fit_transform(df[col])
        df[col].astype(int)
    except:
        pass
```

```python
li = ['Gender','Married','Dependents','Education','Self_Employed','Property_Area']
for i in li:
    trans(X_train,i)
```

```python
X_train.head()
```

```python
Y_train = train_new.iloc[:,-1:]
Y_train.head()
```

```python
model = lr.fit(X_train,Y_train)
```

```python
test = pd.read_csv('test.csv')
test.drop(['Loan_ID'],axis=1,inplace=True)
test.head()
```

```python
test['Gender'].fillna(test['Gender'].mode()[0],inplace=True)
test['Married'].fillna(test['Married'].mode()[0],inplace=True)
test['Dependents'].fillna(test['Dependents'].mode()[0],inplace=True)
test['Self_Employed'].fillna(test['Self_Employed'].mode()[0],inplace=True)
test['Credit_History'].fillna(test['Credit_History'].mode()[0],inplace=True)
test['LoanAmount'].fillna(test['LoanAmount'].mean(),inplace=True)
test['Loan_Amount_Term'].fillna(test['Loan_Amount_Term'].mean(),inplace=True)
```

```python
li = ['Gender','Married','Dependents','Education','Self_Employed','Property_Area']
for i in li:
    trans(test,i)
```

```python
test.head()
```

```python
y_pred = pd.DataFrame(model.predict(test),columns=['Loan_Status'])
```

```python
y_pred.head()
```

## Lets append the prediction to the test file

```python
test['Loan_Status'] = y_pred['Loan_Status']
```
